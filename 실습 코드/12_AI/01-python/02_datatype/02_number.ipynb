# 숫자형
- 정수형
- 실수형

```python
a = 100
b = 123.456

print(a, type(a))
print(b, type(b))
```

```python
# 산술연산
m = 10
n = 3

print(m + n)
print(m - n)
print(m * n)
print(m / n)

print(m // n) # 몫
print(m % n) # 나머지

print(m ** 2)
print(m ** 3)
```

```python
# 반올림/올림/내림(버림)
n = 7 / 15

print(n)

# round함수
print(round(n))
print(round(n, 1))
print(round(n, 2))
```

```python
# 외부 math 모듈(.py)에서 제공되는 함수 사용
import math

print(math.ceil(n))
# print(math.ceil(n, 1))
print(math.ceil(n * 10) / 10) # 0.466 -> 4.66 -> 5.0 -> 0.5

print(math.floor(3.9))
```
