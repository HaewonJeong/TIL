# indexing | slicing

시퀀스자료형에 대해 지원하는 기능

```python
# indexing
text = "hello world"
print(text[0])
print(text[10])
# print(text[11])

print(len(text))

print(text[-1])
print(text[-11])
```

```python
# slicing
# [start:end:step]  start번지부터 end번지전까지 반환

text = "hello world"
print(text[1:4])
print(text[6:9])
print(text[:4])  # 기본값: 0
print(text[4:])  # 기본값: len()
print(text[:])
print(text[6::1]) # [시작 숫자 :: step!]
print(text[6::2])
print(text[::-1])
print(text[6::-1])
```

```python
alpha = 'abcdefg'
print(alpha[::2]) # aceg
print(alpha[1::2])   # bdf
print(alpha[::-1])    # gfedcba
```

# str api

str 자료형이 제공하는 함수(메소드)

```python
# 문자열 길이 확인
print(len('송파 여성인력 개발센터'))

# 포함여부 in연산자
print('AI' in 'AI Agent')
print('ai' in 'AI Agent')

# 동등비교
print("AI" == "AI")
```

```python
# str.replace()
enroll_date = '2026/07/13'

print(enroll_date.replace('/', '-'))  # out-of-place연산
print(enroll_date)
```

```python
# str.strip()

data = '          안녕하세요           반갑습니다.          '
print(data)
print(data.strip())

# 대소문자 관련
original = "hELLO wORLD! 123"
print(original.upper())
print(original.lower())
print(original.capitalize())
print(original.swapcase())
print(original.title())
```
