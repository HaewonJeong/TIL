# tuple

시퀀스객체이면서 불변(immutable)의 객체 순서를 저장

```python
t1 = ()   # 빈 튜플
print(t1, type(t1))

t2 = (10,)
print(t2, type(t2))

# t3 = (10, 20)
t3 = 10, 20
print(t3, type(t3))
```

```python
# 값 치환
x = 33
y = 55

# temp = x
# x = y
# y = temp

x, y = y, x

print(f'x= {x}, y= {y}')
```

```python
foods = ('🍩','🎂','🍬','🍭','🎂')

for food in foods:
    print(food)

# enumerate() 인덱스와 함께 요소를 반환
for k, food in enumerate(foods):
    print(k, food)
```
