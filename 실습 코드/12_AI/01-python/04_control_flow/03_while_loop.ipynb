# while 반복문

- 조건식에 의해 반복횟수가 결정되는 반복문
- 조건식이 True면 반복수행, False면 반복중지
- 반복횟수를 사용자가 결정하는 경우 등에 주로 사용

```python
# while문의 반복횟수를 결정하는 변수 : 증감 변수
i = 0 # 초기식
while i < 3 : #조건식
    print('🍉')
    i += 1 #증감식
```

```python
j = 0
while True: 
    print('🍇')
    j += 1

    if j >= 3:
        break # 현재 반복문을 즉시 중지
```
