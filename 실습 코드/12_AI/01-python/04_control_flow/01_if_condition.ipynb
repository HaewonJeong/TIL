# if 조건문

> 파이썬에서는 들여쓰기로 블럭을 구성한다.

1. if 블럭: 선택적으로 처리하는 경우(해당하는 경우가 존재함)
2. if else 블럭: 참/거짓 둘 중 하나는 무조건 해당되는 경우
3. if elif 블럭: 여러 조건식을 비교해야 하는 경우

# if 블럭

```python
# 점수를 입력받고, 90점 이상인 학생에게만 선물을 전달
score = int(input('점수를 입력해주세요: '))
print(score)

# 90점 이상인 경우만 True가 나올수 있는 조건식을 작성
if score >= 90:
    print('수고하셨습니다. 90점이상 우수학생에게는 🍩🍩🍩🍩를 드립니다.')
```

```python
weekday = input('오늘의 요일을 입력하세요...')
if weekday == '월요일':
    print('월요병이 도졌습니다.')
```

# if else 블럭

```python
# 비밀번호 입력 피드백
password = '1234'

try_password = input('비밀번호를 입력하세요: ')

if password == try_password:
    print('비밀번호가 정확합니다.')
else:
    print('비밀번호가 일치하지 않습니다.')
```

```python
# 입력한 점수가 유효한 경우(0~100)와 아닌 경우
score = int(input('시험점수: '))

# if 0 <= score <= 100:
if 0 <= score and score <= 100:
    if score >= 60:
        print('합격입니다. 🎉🎉🎉🎉')
    else:
        print('불합격입니다. 🥲🥲🥲')
else:
    print('유효하지 않은 점수입니다.')
```

# if elif[else] 블럭

3 가지 이상의 경우의 수가 존재하고, 비교할 조건식도 두 개 이상인 경우

```python
n = int(input('정수 입력: '))

if n > 0:
    print('양수')
elif n == 0:
    print('0')
else:
    print('음수')
```

```python
score = int(input('점수입력: '))

if not(0 <= score <= 100):
    print('유효하지 않은 점수를 입력했습니다.')
else:
    grade = 'F'

    if 90 <= score <= 100:
        grade = 'A'
    elif 80 <= score:
        grade = 'B'
    elif 70 <= score:
        grade = 'C'
    elif 60 <= score:
        grade = 'D'
    
    print(grade)
```
