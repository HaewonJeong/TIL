# if 조건문
> 파이썬에서는 들여쓰기로 블럭을 구성한다.

1. if 블럭: 선택적으로 처리하는 경우(해당하는 경우가 존재함)
2. if else 블럭: 참/거짓 둘 중 하나는 무조건 해당되는 경우
3. if elif 블럭: 여러 조건식을 비교해야 하는 경우

# if 블럭

```python
# 점수를 입력받고, 90점 이상인 학생에게만 선물을 전달
score = int(input('점수를 입력해주세요: '))
print(score)

# 90점 이상인 경우만 True가 나올수 있는 조건식 작성
if score >= 90: 
    print('수고하셨습니다. 90점이상 우수학생에게는 🍩🍩🍩🍩를 드립니다.')
```

```python
weekday = input('오늘의 요일을 입력하세요...')
if weekday == '월요일': 
    print('월요병이 도졌습니다.')
```
