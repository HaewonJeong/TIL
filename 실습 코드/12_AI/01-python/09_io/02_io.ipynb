# 입출력 IO 

## open 함수

`open()` 함수는 파이썬에서 **파일을 열기 위한 기본 함수**이다. 파일을 읽거나 쓰기 위해 사용되며, 다양한 모드로 파일을 열 수 있다.


### 기본 문법

```python
f = open('파일경로', '모드', encoding='인코딩')
```

| 인자       | 설명                                      |
|------------|-------------------------------------------|
| 파일경로   | 열고자 하는 파일의 경로 또는 이름          |
| 모드       | 파일을 여는 방법 (읽기, 쓰기 등)           |
| encoding   | 문자 인코딩 (텍스트 파일인 경우 지정 필요) |

---

### 주요 모드 

| 모드 | 의미                    | 파일 없을 때         |
|------|-------------------------|------------------------|
| `'r'`  | 읽기 모드                | 오류 발생 (`FileNotFoundError`) |
| `'w'`  | 쓰기 모드 (덮어쓰기)     | 새로 생성됨             |
| `'a'`  | 추가 모드 (이어쓰기)     | 새로 생성됨             |
| `'x'`  | 배타적 생성 모드         | 이미 존재하면 오류       |
| `'b'`  | 바이너리 모드 (추가)     | 이진 데이터 처리용       |
| `'t'`  | 텍스트 모드 (기본)       | 생략 가능 (`'rt'` = `'r'`) |

> 예: `'rb'` (이진 읽기), `'wt'` (텍스트 쓰기), `'a'` (텍스트 추가)


### 추가 팁

- `read()` : 전체 내용 문자열로 읽기
- `readline()` : 한 줄씩 읽기
- `readlines()` : 모든 줄을 리스트로 반환
- `write()` : 문자열 1개 작성
- `writelines()` : 리스트 형식 문자열 여러 개 작성
